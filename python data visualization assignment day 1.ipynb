```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#question 1
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,'bo-',linewidth=2,markersize=8)
plt.title("LINE PLOT",color='r')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show()
```

```python
#question 2
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180]
sales_2 = [70,90,160,150,140,145,175]
plt.plot(days,sales_1,'bo-',label="company 1",linewidth=2,markersize=8)
plt.plot(days,sales_2,'ro-',label="company 2",linewidth=2,markersize=8)
plt.title("SALES TREND BETWEEN COMPANY 1 AND 2")
plt.xlabel("DAYS")
plt.ylabel("SALES")
plt.legend()
plt.show()

```

```python
#question 3
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
plt.subplot(3,3,1)
plt.plot(x,y1,'bo-')
plt.subplot(3,3,2)
plt.plot(x,y2,'ro-')
plt.subplot(3,3,3)
plt.plot(x,y3,'yo-')
plt.subplot(3,3,4)
plt.plot(x,y4,'go-')
plt.subplot(3,3,5)
plt.plot(x,y5,'bo--')
plt.show()
```
